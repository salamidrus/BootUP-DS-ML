# Web Scraping with Beautifulsoup

- pip install bs4
- pip install requests
- pip install html5lib

```python
import requests
import bs4
```

```python
url = "https://bootup.ai/blog/belajar-laravel-dan-tutorialnya-untuk-pemula-lengkap/"
r = requests.get(url)
print(r.content)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.content, 'html5lib')
print(soup.prettify())
```

```python
table = soup.find('div',attrs={'class':'container'})
for row in table.findAll('div'):
    print(row.a['href'])
    print(row.p)
```
